```python
# read in and extract data for ticer BAC (BankOfAm)
f = h5py.File('./Data/Project-1-Prediction/Book_event_levelII/book_events_sip-nyse_2016-05-02.h5', 'r')
data = f['BAC'][:] # ticker BAC
data_ticker = data
f.close()

print(data_ticker[0:3]) # first 3 rows of the ticker data
```

```python
df1 = pd.DataFrame(data_ticker) # convert to dataframe structure
df1.shape # data for BAC is 410k
```

```python
df1.head()
```

```python
# formatting columns

df1['book_event_type'] = df1['book_event_type'].astype(str) # convert datatype for extraction op
df1['side'] = df1['side'].astype(str)
df1['book_event_type_extr'] = df1['book_event_type'].str[2:3] # extract main character
df1['side_extr'] = df1['side'].str[2:3]
df1['timestamp_dt'] = pd.to_datetime(df1['timestamp']).dt.tz_localize('UTC').dt.tz_convert('US/Eastern')  # convert to readable timestamp + ET timezone
df1.drop(columns = ["book_event_type","side","aux_quantity","aux1","aux2"],inplace = True) # drop unneccesary columns
df1 = df1[['order_id', 'price', 'quantity', 'book_event_type_extr', 'side_extr','timestamp_dt','timestamp']] # rearrange columns
df1.columns = ['order_id', 'price', 'qty', 'book_event_type', 'side','timestamp_dt','timestamp'] # rename
df1 = df1.sort_values(by = ['timestamp_dt','side'],ascending = True)
df1.head()
```

#### Visuals to get an idea of the health of the data
(data - ticker "BAC" only)

```python
# for plotting
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import pylab
import matplotlib.dates as mdates
```

```python
df2 = df1 # will leave df1 as original dataframe
```

```python
# extract date and hour of the day for plots
import datetime
df2['month'] = df2['timestamp_dt'].dt.month
df2['day'] = df2['timestamp_dt'].dt.day
df2['hour'] = df2['timestamp_dt'].dt.hour
df2['date_minute'] = df2['timestamp_dt'].apply(lambda dt: datetime.datetime(dt.year, dt.month, dt.day, dt.hour,dt.minute))
df2.head()
```

##### Total event counts in the data 

```python
# BY THE HOUR 
sns.set_style("darkgrid")
fig = plt.figure(figsize=(20,8))
ax = fig.add_subplot(111)

df_plot = df2.groupby(by=['month','day','hour']).size().reset_index(name="counts")
df_plot['month']=df_plot['month'].apply(lambda x: '{0:0>2}'.format(x))
df_plot['day']=df_plot['day'].apply(lambda x: '{0:0>2}'.format(x))
df_plot['hour']=df_plot['hour'].apply(lambda x: '{0:0>2}'.format(x))
df_plot['day_hour'] = df_plot["month"].map(str) + "-" + df_plot["day"].map(str) + " " + df_plot["hour"].map(str) 
df_plot = df_plot.sort_values(by=['day_hour'])
#df_plot.head()

# plot the trends
plt.plot('day_hour', 'counts', data=df_plot, marker='o', color='mediumvioletred')

# format plots
fig.suptitle('Book Events by the hour for BAC', fontsize=20)
plt.xlabel('Time (Format: MM-DD hr)', fontsize=16)
plt.ylabel('Event Count', fontsize=16)
pylab.legend(loc='upper right')
ax.get_yaxis().set_major_formatter(
    matplotlib.ticker.FuncFormatter(lambda y, p: format(int(y), ',')))

# display
plt.show()

# trading stars at 9am and finishes by 4pm #verified
```

```python
# BY THE MINUTE
sns.set_style("darkgrid")
fig = plt.figure(figsize=(20,8))
ax = fig.add_subplot(111)

df_plot = df2.groupby(by=['date_minute']).size().reset_index(name="counts")
df_plot = df_plot.sort_values(by=['date_minute'])
#df_plot.head()

# plot the trends
plt.plot('date_minute', 'counts', data = df_plot, color='mediumvioletred')

# format plots
fig.suptitle('Book Events by the minute for BAC', fontsize=20)
plt.xlabel('Time (Format: MM-DD hr)', fontsize=16)
plt.ylabel('Event Count', fontsize=16)
pylab.legend(loc='upper right')
ax.get_yaxis().set_major_formatter(
    matplotlib.ticker.FuncFormatter(lambda y, p: format(int(y), ',')))

# display
plt.show()
```

##### Ask (A) /Bid (B) /Traded (U) book events counts 

```python
# BY THE HOUR 
sns.set_style("darkgrid")
fig = plt.figure(figsize=(20,8))
ax = fig.add_subplot(111)

df_plot = df2.groupby(by=['month','day','hour','side']).size().reset_index(name="counts")
df_plot['month']=df_plot['month'].apply(lambda x: '{0:0>2}'.format(x))
df_plot['day']=df_plot['day'].apply(lambda x: '{0:0>2}'.format(x))
df_plot['hour']=df_plot['hour'].apply(lambda x: '{0:0>2}'.format(x))
df_plot['day_hour'] = df_plot["month"].map(str) + "-" + df_plot["day"].map(str) + " " + df_plot["hour"].map(str) 
df_plot = df_plot.sort_values(by=['day_hour'])
df_plot = df_plot.pivot(index='day_hour', columns='side', values='counts')
df_plot = df_plot.reset_index()
#df_plot.head()

# plot the trends
plt.plot('day_hour', 'A', data=df_plot, marker='o',color='tomato')
plt.plot('day_hour', 'B', data=df_plot, marker='o', color='deepskyblue')
plt.plot('day_hour', 'U', data=df_plot, marker='o',label="U or Traded", color='indigo')


# format plots
fig.suptitle('Book Events by the hour for BAC', fontsize=20)
plt.xlabel('Time (Format: MM-DD hr)', fontsize=16)
plt.ylabel('Event Count', fontsize=16)
pylab.legend(loc='upper right')
ax.get_yaxis().set_major_formatter(
    matplotlib.ticker.FuncFormatter(lambda y, p: format(int(y), ',')))

# display
plt.show()

# roughly equal spread of bid vs ask - ask geenrally high implying high liquidity
```

```python
# BY THE MINUTE
fig = plt.figure(figsize=(20,8))
ax = fig.add_subplot(111)
sns.set_style("darkgrid")

df_plot = df2.groupby(by=['date_minute','side']).size().reset_index(name="counts")
df_plot = df_plot.sort_values(by=['date_minute'])
df_plot = df_plot.pivot(index='date_minute', columns='side', values='counts')
df_plot = df_plot.reset_index()
#df_plot.head()

# plot the trends
plt.plot('date_minute', 'A', data=df_plot,color='tomato')
plt.plot('date_minute', 'B', data=df_plot, color='deepskyblue')
plt.plot('date_minute', 'U', data=df_plot,label="U or Traded", color='indigo')


# format plots
fig.suptitle('Bid/Ask Book Event Count by the minute for BAC', fontsize=20)
plt.xlabel('Time (Format: MM-DD hr)', fontsize=16)
plt.ylabel('Event Count', fontsize=16)
pylab.legend(loc='upper right')
ax.get_yaxis().set_major_formatter(
    matplotlib.ticker.FuncFormatter(lambda y, p: format(int(y), ',')))


# display
plt.show()
```
